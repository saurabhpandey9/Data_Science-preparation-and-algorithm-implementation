# Plotly Module (for web interface of charts)

### Interactive chart using plotly

```python
#plotly module needed to be install in anaconda to use it
```

```python
import plotly
import plotly.graph_objs as go
```

```python
plotly.offline.plot({
    'data':[go.Scatter(x=data['5.1'].tolist(),y=data["1.4"])],
    "layout" :go.Layout(title='Hello Saurabh')
},auto_open=True)
#here tolist() nahi bhi likoge to chalega ,same result produce karega
```

```python
#writing in more readable manner
df=go.Scatter(x=data['5.1'],y=['1.4'],
             mode='markers',
             name='markers')
layout=go.Layout(title='practce',showlegend=True)

plotly.offline.plot({'data':[df],'layout':layout},auto_open=True)
```

### Makers and lines in one figure

```python
df=go.Scatter(x=data['5.1'],y=['1.4'],
             mode='markers',
             name='markers')
df2=go.Scatter(x=data['5.1'],y=['1.4'],
             mode='lines+markers',
             name='lines+markers')
#here mode cane lines,markers, lines+markers
layout=go.Layout(title='practce',showlegend=True)

plotly.offline.plot({'data':[df,df2],'layout':layout},auto_open=True)
```
